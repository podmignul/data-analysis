```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import seaborn as sns
from collections import Counter
```

Часть 1

```python
# Import Data
diamonds_data = pd.read_csv('diamonds.csv')
diamonds_data
```

График №1

```python
dat = diamonds_data['depth']

bs = np.arange(min(dat) + 5, max(dat) - 5)

j = diamonds_data.loc[diamonds_data['color'] == 'J']
i = diamonds_data.loc[diamonds_data['color'] == 'I']
h = diamonds_data.loc[diamonds_data['color'] == 'H']
g = diamonds_data.loc[diamonds_data['color'] == 'G']
e = diamonds_data.loc[diamonds_data['color'] == 'E']
f = diamonds_data.loc[diamonds_data['color'] == 'F']
d = diamonds_data.loc[diamonds_data['color'] == 'D']

plt.hist(d['depth'], bs, label = 'D', color = '#420951')
plt.hist(e['depth'], bs, label = 'E', color = '#423a81')
plt.hist(f['depth'], bs, label = 'F', color = '#2d698d')
plt.hist(g['depth'], bs, label = 'G', color = '#22918d')
plt.hist(h['depth'], bs, label = 'H', color = '#38b77d')
plt.hist(i['depth'], bs, label = 'I', color = '#92d652')
plt.hist(j['depth'], bs, label = 'J', color = '#fde643')


plt.legend(title = 'color')
plt.xlabel("depth")
plt.show()
```

```python
# data
data = mtcars_data.loc[mtcars_data.am == 0]
dat = mtcars_data.loc[mtcars_data.am == 1]

am = mtcars_data['am']
cyl = mtcars_data['cyl']
# use the scatterplot function to build the bubble map
sns.scatterplot(data=data, x="mpg", y="hp", hue = am, size=cyl, legend=False, sizes=(40, 200))
sns.scatterplot(data=dat, x="mpg", y="hp", hue = am, size=cyl, sizes=(40, 200))

# show the graph
plt.show()
```

Часть 3

Данные по углекислому газу EPICA Dome C Ice Core 800KYr
Описание:
Запись углекислого газа из EPICA (Европейский проект по отбору керна из льда в Антарктиде) Купол C ледяной керн, охватывающий от 0 до 800 тыс. лет н.э.
Формат:
Фрейм данных с 1096 наблюдениями по следующим 2 переменными: 
1) возраст(Возраст в годах до настоящего времени (АД))
2) co2(Уровень CO2 (ppm))

```python
# Import Data
edcCO2_data = pd.read_csv('edcCO2.csv')
edcCO2_data
```

Диаграмма 1

#Данные
data = Counter(edcCO2_data['co2'])
labels = data.keys()
sizes = data.values()

#Круговая диаграмма
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Распределение уровня CO2 с возрастом')

plt.show()

Диаграмма 2

```python
sns.histplot(edcCO2_data,x="co2")
plt.title("Гистограмма для признака co2")

plt.show()
```

Диаграмма 3

```python
#Точечная диаграмма
sns.scatterplot(data=edcCO2_data, x="co2", y="age", color='orange')
plt.title('Точечная диаграмма по признакам co2 и age')

plt.show()
```

Диаграмма 4

```python
#Ящик с усами
sns.boxplot(data=edcCO2_data, x="co2")
plt.title("Зависимость температуры от месяца")

plt.show()
```

Диаграмма 5

```python
#Данные
data = edcCO2_data.corr()

#Тепловая карта

sns.heatmap(data,annot=True, fmt=".2f")
plt.title('Корреляция признаков')

plt.show()
```

Диаграмма 6

```python
# data
fig, ax = plt.subplots()

x = edcCO2_data['co2']
y = edcCO2_data['age']

ax.plot(x, y,  color='red', linestyle='--', marker='o', alpha=0.2)

plt.title("Зависимость co2 от возраста")
plt.xlabel("co2")

plt.show()
```

Диаграмма 7

```python

co2 = edcCO2_data['co2']
age = edcCO2_data['age']
ts = pd.Series(co2, index = age) 

ts = ts.cumsum() 
plt.title("Последовательность увеличения возраста")
ts.plot() 
plt.show()
```
