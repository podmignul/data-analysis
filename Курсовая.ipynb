```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import seaborn as sns
sns.set()
```

```python
# Import Data
mh = pd.read_csv('Mlbb_Heroes.csv')
mh
```

## Роль

Для выбора наиболее подходящего героя, стоит проанализировать имеющиеся данные и составить подходящие диграммы. 
Были задействованы десять диаграмм, в совокупности они представляют собой четыре разных вида анализа данных (круговая диаграмма, гистограмма, точечная диграмма, столбчатая диаграмма).   

1. Основная роль героя, с какой частотой встречается та или иная роль, круговая диаграмма

```python
#Данные
data = Counter(mh['Primary_Role'])
labels = data.keys()
sizes = data.values()

#Круговая диаграмма
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90,
 colors = ['#4FB0C6','#F0D879','#739e3b', '#ec693f', '#e9bba9', '#a47c64' ])

plt.show()
```

Исходя из диграммы 1 видно, что у большей части игровых персонажей основной ролью является "Боец", на втором месте "Маг", на третьем "Стрелок", на четвётром "Танк", на последних местях располагаются роль "Убийца" и "Поддержка". Данный результат довольно интересен с точки зрения подбора своего персонажа, так как получается, что изначально при выборе игрока глаз падает на самые крупные категории, соответсвенно шанс, что именно среди роли "Боец" или "Маг" человек сможет подобрать своего героя в два раза больше, чем среди оставшихся ролей.  

2. У какого количества героев имеется второстепенная роль, вывести их имена, можно вывести также основные роли, посмотреть если ли взаимосвязь

```python
#Данные
data = Counter(mh['Secondary_Role'])
labels = data.keys()
sizes = data.values()

#Круговая диаграмма
plt.pie(sizes, labels = labels, colors = ['#bac2c6','#e9bba9','#F0D879','#4FB0C6', '#ec693f', '#739e3b', '#a47c64' ],
 autopct = '%1.0f%%', pctdistance = 0.8, rotatelabels = True)

plt.show()
```

```python
#Гистограмма
sns.histplot(mh,x ='Secondary_Role', color = '#fde643')

plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.show()
```

## Сравнение маны и здоровья

6. Сравнение количества маны и здоровья у определённых героев

```python
# Столбчатая диграмма
sales_data = mh.groupby('Primary_Role')[['Hp', 'Mana']].mean()

plt.bar(sales_data.index, sales_data['Hp'], color = '#1ee3cf', label = 'Hp')
plt.bar(sales_data.index, sales_data['Mana'], color = 'b', label = 'Mana')

plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.legend(loc = 'center')

plt.show()
```
